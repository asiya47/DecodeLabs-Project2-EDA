```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("All libraries imported successfully!")
```

```python
df["CouponCode"] = df["CouponCode"].fillna("No Coupon")
```

```python
df.isnull().sum()
```

```python
#Product Analysis
df["Product"].value_counts()
```

```python
#Then visualize it
plt.figure(figsize=(10,5))
sns.countplot(data=df, y="Product", order=df["Product"].value_counts().index)
plt.title("Product Distribution")
plt.xlabel("Number of Orders")
plt.ylabel("Product")
plt.show()
```

```python
#Payment Method Analysis
df["PaymentMethod"].value_counts()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="PaymentMethod")
plt.title("Payment Method Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
#Order Status Analysis
df["OrderStatus"].value_counts()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="OrderStatus")
plt.title("Order Status Distribution")
plt.show()
```

```python
#Referral Source
df["ReferralSource"].value_counts()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(data=df, x="ReferralSource")
plt.title("Referral Sources")
plt.xticks(rotation=45)
plt.show()
```

```python
#Coupon Usage
df["CouponCode"].value_counts()
```

```python
#6. Monthly Sales Trend
df["Month"] = df["Date"].dt.to_period("M")

monthly_sales = df.groupby("Month")["TotalPrice"].sum()

plt.figure(figsize=(12,5))
monthly_sales.plot(marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

```python
#7. Correlation Heatmap
plt.figure(figsize=(6,5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#8. Outlier Detection
plt.figure(figsize=(7,4))
sns.boxplot(x=df["TotalPrice"])
plt.title("Outliers in Total Price")
plt.show()
```

```python
#9: Which Product Generates the Highest Revenue?
product_sales = df.groupby("Product")["TotalPrice"].sum().sort_values(ascending=False)

print(product_sales)
```

```python
plt.figure(figsize=(10,5))
product_sales.plot(kind="bar", color="skyblue")
plt.title("Revenue by Product")
plt.xlabel("Product")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
#10: Average Order Value by Payment Method
payment_avg = df.groupby("PaymentMethod")["TotalPrice"].mean()

print(payment_avg)
```

```python
plt.figure(figsize=(7,5))
payment_avg.plot(kind="bar", color="orange")
plt.title("Average Order Value by Payment Method")
plt.ylabel("Average Order Value")
plt.xticks(rotation=45)
plt.show()
```

```python
#11: Quantity vs Total Price
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x="Quantity", y="TotalPrice")
plt.title("Quantity vs Total Price")
plt.show()
```

```python
#12: Top 10 Shipping Addresses
top_addresses = df["ShippingAddress"].value_counts().head(10)

print(top_addresses)
```

```python
plt.figure(figsize=(10,5))
top_addresses.plot(kind="bar")
plt.title("Top 10 Shipping Locations")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)
plt.show()
```

```python
#13: Items in Cart Distribution
plt.figure(figsize=(7,5))
sns.histplot(df["ItemsInCart"], bins=10, kde=True)
plt.title("Items in Cart Distribution")
plt.xlabel("Items in Cart")
plt.show()
```

```python
#14: Highest Revenue Month
monthly_sales = df.groupby(df["Date"].dt.to_period("M"))["TotalPrice"].sum()

print(monthly_sales)

highest_month = monthly_sales.idxmax()
highest_sales = monthly_sales.max()

print("Highest Revenue Month:", highest_month)
print("Revenue:", highest_sales)
```

```python
#15: Average Quantity Sold per Product
avg_quantity = df.groupby("Product")["Quantity"].mean().sort_values(ascending=False)

print(avg_quantity)
```

```python
plt.figure(figsize=(10,5))
avg_quantity.plot(kind="bar")
plt.title("Average Quantity Sold per Product")
plt.ylabel("Average Quantity")
plt.xticks(rotation=45)
plt.show()
```

```python
#16: Top 10 Highest Value Orders
top_orders = df.nlargest(10, "TotalPrice")

top_orders[["OrderID", "Product", "TotalPrice"]]
```

```python
#17: Summary Statistics
print("Total Revenue :", df["TotalPrice"].sum())
print("Average Revenue :", df["TotalPrice"].mean())
print("Maximum Revenue :", df["TotalPrice"].max())
print("Minimum Revenue :", df["TotalPrice"].min())
print("Total Orders :", len(df))
```
